## Lab-1: Implementation of XOR Neural Network

```python
import numpy as np
```

```python
# step activation function
def step(z):
    return 1 if z >= 0 else 0
```

```python
# XOR Neural Network
def xor_nn(x1, x2):
    # Hidden layer (OR and NAND)
    h1 = step(1*x1 + 1*x2 - 0.5) #OR
    h2 = step(-1*x1 - 1*x2 + 1.5) #NAND
    # output layer (AND)
    y = step(1*h1 + 1*h2 - 1.5) #XOR
    return y
```

```python
# Testing
print("XOR(0,0) = ", xor_nn(0,0)) #0
print("XOR(0,1) = ", xor_nn(0,1)) #1
print("XOR(1,0) = ", xor_nn(1,0)) #1
print("XOR(1,1) = ", xor_nn(1,1)) #0
```

```python
# XNOR Neural Network
def xnor_nn(x1, x2):
    # Hidden layer (~XOR)
    h1 = step(1*x1 + 1*x2 - 0.5) #OR
    h2 = step(-1*x1 - 1*x2 + 1.5) #NAND
    # output layer (AND)
    y = step((-1*h1) + (-1*h2) + 1.5) #XOR
    return y
```

```python
# Testing
print("XNOR(0,0) = ", xnor_nn(0,0)) #1
print("XNOR(0,1) = ", xnor_nn(0,1)) #0
print("XNOR(1,0) = ", xnor_nn(1,0)) #0
print("XNOR(1,1) = ", xnor_nn(1,1)) #1
```

```python
import torch
import torch.nn as nn
import torch.optim as optim
```

```python
X = torch.tensor([[0., 0.],
                  [0., 1.],
                  [1., 0.],
                  [1., 1.]])

y = torch.tensor([[0.],
                  [1.],
                  [1.],
                  [0.]])
```

```python
class XORNet(nn.Module):
    def __init__(self):
        super().__init__()
        self.hidden = nn.Linear(2, 2)   # 2 inputs → 2 hidden neurons
        self.output = nn.Linear(2, 1)   # 2 hidden → 1 output
        self.sigmoid = nn.Sigmoid()

    def forward(self, x):
        x = self.sigmoid(self.hidden(x))   # hidden layer
        x = self.sigmoid(self.output(x))   # output layer
        return x
```

```python
model = XORNet()
criterion = nn.BCELoss()        # or BCELoss
optimizer = optim.SGD(model.parameters(), lr=0.5)
```

```python
for epoch in range(5000):
    optimizer.zero_grad()       # clear old gradients

    y_pred = model(X)           # forward pass
    loss = criterion(y_pred, y) # compute loss

    loss.backward()             # backpropagation
    optimizer.step()            # update weights

    if epoch % 500 == 0:
        print(f"Epoch {epoch}, Loss = {loss.item():.4f}")
```

```python
with torch.no_grad():
    predictions = model(X)
    print(torch.round(predictions))
```
