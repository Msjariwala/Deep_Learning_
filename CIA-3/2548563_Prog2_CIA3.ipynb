```python
# Data Upload & Preprocessing
Docs = "The India AI Impact Summit 2026, held from February 16вҖ“21, 2026, at Bharat Mandapam, New Delhi, was a major global event focused on AI for social good, governance, and infrastructure. Inaugurated by Prime Minister Narendra Modi, it highlighted India's role as a leading AI hub for the Global South, with significant investments from global tech majors"
```

```python
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
tokenizer = Tokenizer()
```

```python
tokenizer.fit_on_texts([Docs])
```

```python
#Assigned numbers or integer to al the words present in the sentences.
tokenizer.word_index
```

```python
len(tokenizer.word_index)
```

```python
# Now review sentence by sentence and create the integer values in sequence for all the given sentences in input data, For that first we find the each sentence in the dataset.
for sentence in Docs.split(','):
    print(sentence)
    print(tokenizer.texts_to_sequences([sentence])[0])
```

```python
tokenized_sentence = tokenizer.texts_to_sequences([sentence])[0]
for i in range(1,len(tokenized_sentence)):
    print(tokenized_sentence[:i+1])
```

```python
input_sequences =[]
for sentence in Docs.split('.'):
  tokenized_sentence = tokenizer.texts_to_sequences([sentence])[0]
  for i in range(1,len(tokenized_sentence)):
    input_sequences.append(tokenized_sentence[:i+1]) # Collect the sequences
    print(tokenized_sentence[:i+1])
```

```python
# calculate the length of the maximum sequence present.
maximum_length = max([len(x) for x in input_sequences])
```

```python
# Do padding
from tensorflow.keras.utils import pad_sequences
padded_sequences = pad_sequences(input_sequences, maxlen = maximum_length, padding ='pre')
display(padded_sequences)
```

```python
# Now split the input and output from the padded sequences
#now input of padded sequences will have all the word integers except last one in each row.
X = padded_sequences[:,:-1]
display(X)
```

```python
#Similarly, for the output
y = padded_sequences[:, -1]
display(y)
```

```python
# Now, Predicting the output wil be in classification task. So, converting the integer values in One hot encoded vector.
X.shape
```

```python
y.shape
```

```python
from tensorflow.keras.utils import to_categorical
to_categorical(y, num_classes = len(tokenizer.word_index)+1)
```

```python
y = to_categorical(y, num_classes = len(tokenizer.word_index)+1)
y.shape
```

```python
#Architecture of LSTM
#THREE LAYERS
# 1. Embedding layer 2. LSTM Layer 3. Dense Layer
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense
```

```python
model = Sequential()
model.add(Embedding(len(tokenizer.word_index)+1, 10))
model.add(LSTM(100))
model.add(Dense(len(tokenizer.word_index)+1, activation = 'softmax'))
```

```python
model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
model.build(input_shape=(None, maximum_length-1)) # (batch_size, input_length)
model.summary()
```

```python
model.fit(X,y,epochs=150)
```

```python
import numpy as np

text = "significant"
# tokenize
token_text = tokenizer.texts_to_sequences([text])[0]
#padding
padded_token_text = pad_sequences([token_text], maxlen=maximum_length-1, padding='pre')
#predict
position = np.argmax(model.predict(padded_token_text))

for word,index in tokenizer.word_index.items():
    if index == position:
        print(word)
```
